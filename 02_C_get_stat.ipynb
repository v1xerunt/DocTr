```python
import pickle
import numpy as np
```

```python
trial_phase = pickle.load(open('./data/trial_phase.pkl', 'rb'))
trial_info = pickle.load(open('./data/trial_info.pkl', 'rb'))
```

```python
# get phase stat
phase_stat = {}
for trial in trial_phase:
    if trial_phase[trial] is None or 'Phase' not in trial_phase[trial]:
            phase_stat['None'] = phase_stat.get('None', 0) + 1
            continue
    if 'Phase 1' in trial_phase[trial]:
        phase_stat['Phase 1'] = phase_stat.get('Phase 1', 0) + 1
    if 'Phase 2' in trial_phase[trial]:
        phase_stat['Phase 2'] = phase_stat.get('Phase 2', 0) + 1
    if 'Phase 3' in trial_phase[trial]:
        phase_stat['Phase 3'] = phase_stat.get('Phase 3', 0) + 1
    if 'Phase 4' in trial_phase[trial]:
        phase_stat['Phase 4'] = phase_stat.get('Phase 4', 0) + 1
print(phase_stat)
```

```python
for each_trial in trial_info:
    if str(trial_info[each_trial]['start_date']) == 'nan':
        print(each_trial)
        continue
```

```python
'''
    Structure of payment_dict:
    Key: Trial NCTID
    Value: {
        [
            {
                'NPI': Amount, ...
            },
            {
                'Hospital ID': {NPI: Amount, ...}
            },
            {
                'Entity name': {NPI: Amount, ...}
            }
        ]
'''

payment_dict = pickle.load(open('../data/payment_dict.pkl', 'rb'))
```

```python
payment_dict
```

```python
# Get the total payment for each trial
trial_payment = {}
for trial in payment_dict:
    trial_payment[trial] = []
    for payment in payment_dict[trial][0]:
        trial_payment[trial].append(payment_dict[trial][0][payment])
    if len(payment_dict[trial][1]) > 0:
        for payment in payment_dict[trial][1]:
            trial_payment[trial].append(max(payment_dict[trial][1][payment].values()))
    if len(payment_dict[trial][2]) > 0:
        for payment in payment_dict[trial][2]:
            trial_payment[trial].append(max(payment_dict[trial][2][payment].values()))
        
total_payment = {}
for trial in trial_payment:
    total_payment[trial] = np.sum(trial_payment[trial])
    
# Print statistics of total payment
print('Total payment statistics:')
print('Mean: ', np.mean(list(total_payment.values())))
print('std: ', np.std(list(total_payment.values())))
print('Median: ', np.median(list(total_payment.values())))
print('Max: ', np.max(list(total_payment.values())))
print('Min: ', np.min(list(total_payment.values())))
```

```python
# Get the npi list for each trial
trial_npi = {}
for trial in payment_dict:
    trial_npi[trial] = []
    for payment in payment_dict[trial][0]:
        trial_npi[trial].append(payment)
    if len(payment_dict[trial][1]) > 0:
        for payment in payment_dict[trial][1]:
            trial_npi[trial] += list(payment_dict[trial][1][payment].keys())
    if len(payment_dict[trial][2]) > 0:
        for payment in payment_dict[trial][2]:
            trial_npi[trial] += list(payment_dict[trial][2][payment].keys())
    trial_npi[trial] = list(set(trial_npi[trial]))

# Print statistics of npi list
print('NPI list statistics:')
print('Mean: ', np.mean([len(trial_npi[trial]) for trial in trial_npi]))
print('std: ', np.std([len(trial_npi[trial]) for trial in trial_npi]))
print('Median: ', np.median([len(trial_npi[trial]) for trial in trial_npi]))
print('Max: ', np.max([len(trial_npi[trial]) for trial in trial_npi]))
print('Min: ', np.min([len(trial_npi[trial]) for trial in trial_npi]))
```

```python
trial_hos = {}
for trial in payment_dict:
    trial_hos[trial] = []
    if len(payment_dict[trial][1]) > 0:
        for payment in payment_dict[trial][1]:
            trial_hos[trial] += list(payment_dict[trial][1][payment].keys())
    trial_hos[trial] = list(set(trial_hos[trial]))
    
trial_entity = {}
for trial in payment_dict:
    trial_entity[trial] = []
    if len(payment_dict[trial][2]) > 0:
        for payment in payment_dict[trial][2]:
            trial_entity[trial] += list(payment_dict[trial][2][payment].keys())
    trial_entity[trial] = list(set(trial_entity[trial]))

print('Total number of trials: ', len(trial_npi))
print('Total number of NPIs: ', len(set([npi for trial in trial_npi for npi in trial_npi[trial]])))
print('Total number of hospitals', len(set([hos for trial in trial_hos for hos in trial_hos[trial]])))
print('Total number of entities', len(set([entity for trial in trial_entity for entity in trial_entity[trial]])))
```
