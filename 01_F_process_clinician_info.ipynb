```python
import pandas as pd
import numpy as np
import pickle
```

```python
npi_list = pickle.load(open('./data/filtered_npi.pkl', 'rb'))
npi_list = [str(npi) for npi in npi_list]
```

```python
data_dict = {}


for i in range(2017, 2022):
    payment = pd.read_csv('D:/Data/payment/Research/OP_DTL_RSRCH_PGYR' + str(i) + '_P01202023.csv', dtype={'Covered_Recipient_NPI':str, 'Principal_Investigator_1_NPI': str, 'Principal_Investigator_2_NPI': str, 'Principal_Investigator_3_NPI': str, 'Principal_Investigator_4_NPI': str, 'Principal_Investigator_5_NPI': str, 'ClinicalTrials_Gov_Identifier': str})

    payment = payment[(payment['Total_Amount_of_Payment_USDollars'] > 0) & payment['ClinicalTrials_Gov_Identifier'].notna()]
    institution_payment = payment[(payment['Principal_Investigator_1_NPI'].isin(npi_list) | payment['Principal_Investigator_2_NPI'].isin(npi_list) | payment['Principal_Investigator_3_NPI'].isin(npi_list) | payment['Principal_Investigator_4_NPI'].isin(npi_list) | payment['Principal_Investigator_5_NPI'].isin(npi_list)) & payment['Teaching_Hospital_ID'].notnull() & payment['Noncovered_Recipient_Entity_Name'].isnull()]
    entity_payment = payment[(payment['Principal_Investigator_1_NPI'].isin(npi_list) | payment['Principal_Investigator_2_NPI'].isin(npi_list) | payment['Principal_Investigator_3_NPI'].isin(npi_list) | payment['Principal_Investigator_4_NPI'].isin(npi_list) | payment['Principal_Investigator_5_NPI'].isin(npi_list)) & payment['Noncovered_Recipient_Entity_Name'].notnull() & payment['Teaching_Hospital_ID'].isnull()]
    individual_payment = payment[payment['Covered_Recipient_NPI'].isin(npi_list)]
    physician_payment = payment[payment['Physician_NPI'].isin(npi_list)]

    for idx, row in individual_payment.iterrows():
        cur_nct = row['ClinicalTrials_Gov_Identifier']
        cur_npi = row['Covered_Recipient_NPI']
        
        if cur_npi not in data_dict:
            data_dict[cur_npi] = {}
            data_dict[cur_npi]['First_Name'] = row['Covered_Recipient_First_Name']
            data_dict[cur_npi]['Middle_Name'] = row['Covered_Recipient_Middle_Name']
            data_dict[cur_npi]['Last_Name'] = row['Covered_Recipient_Last_Name']
            data_dict[cur_npi]['Name_Suffix'] = row['Covered_Recipient_Name_Suffix']
            data_dict[cur_npi]['Primary_Business_Street_Address_Line1'] = row['Recipient_Primary_Business_Street_Address_Line1']
            data_dict[cur_npi]['Primary_Business_Street_Address_Line2'] = row['Recipient_Primary_Business_Street_Address_Line2']
            data_dict[cur_npi]['City'] = row['Recipient_City']
            data_dict[cur_npi]['State'] = row['Recipient_State']
            data_dict[cur_npi]['Zip_Code'] = row['Recipient_Zip_Code']
            data_dict[cur_npi]['Country'] = row['Recipient_Country']
            data_dict[cur_npi]['Province'] = row['Recipient_Province']
            data_dict[cur_npi]['Postal_Code'] = row['Recipient_Postal_Code']
            for i in range(1,7):
                data_dict[cur_npi]['Primary_Type_' + str(i)] = row['Covered_Recipient_Primary_Type_' + str(i)]
            for i in range(1,7):
                data_dict[cur_npi]['Specialty_' + str(i)] = row['Covered_Recipient_Specialty_' + str(i)]
        
    for idx, row in institution_payment.iterrows():
        cur_nct = row['ClinicalTrials_Gov_Identifier']
        cur_hos = row['Teaching_Hospital_ID']
        
        for i in range(1, 6):
            if str(row['Principal_Investigator_' + str(i) + '_NPI']) == 'nan':
                continue
            cur_npi = row['Principal_Investigator_' + str(i) + '_NPI']
            if cur_npi not in data_dict:
                data_dict[cur_npi] = {}
                data_dict[cur_npi]['First_Name'] = row['Principal_Investigator_' + str(i) + '_First_Name']
                data_dict[cur_npi]['Middle_Name'] = row['Principal_Investigator_' + str(i) + '_Middle_Name']
                data_dict[cur_npi]['Last_Name'] = row['Principal_Investigator_' + str(i) + '_Last_Name']
                data_dict[cur_npi]['Name_Suffix'] = row['Principal_Investigator_' + str(i) + '_Name_Suffix']
                data_dict[cur_npi]['Primary_Business_Street_Address_Line1'] = row['Principal_Investigator_' + str(i) + '_Business_Street_Address_Line1']
                data_dict[cur_npi]['Primary_Business_Street_Address_Line2'] = row['Principal_Investigator_' + str(i) + '_Business_Street_Address_Line2']
                data_dict[cur_npi]['City'] = row['Recipient_City']
                data_dict[cur_npi]['State'] = row['Recipient_State']
                data_dict[cur_npi]['Zip_Code'] = row['Recipient_Zip_Code']
                data_dict[cur_npi]['Country'] = row['Recipient_Country']
                data_dict[cur_npi]['Province'] = row['Recipient_Province']
                data_dict[cur_npi]['Postal_Code'] = row['Recipient_Postal_Code']
                for j in range(1,7):
                    data_dict[cur_npi]['Primary_Type_' + str(j)] = row['Principal_Investigator_' + str(i) + '_Primary_Type_'+ str(j)]
                for j in range(1,7):
                    data_dict[cur_npi]['Specialty_' + str(j)] = row['Principal_Investigator_' + str(i) + '_Specialty_' + str(j)]               
                

    for idx, row in entity_payment.iterrows():
        cur_nct = row['ClinicalTrials_Gov_Identifier']
        cur_ent = row['Noncovered_Recipient_Entity_Name']
        
        for i in range(1, 6):
            if str(row['Principal_Investigator_' + str(i) + '_NPI']) == 'nan':
                continue
            cur_npi = row['Principal_Investigator_' + str(i) + '_NPI']
            if cur_npi not in data_dict:
                data_dict[cur_npi] = {}
                data_dict[cur_npi]['First_Name'] = row['Principal_Investigator_' + str(i) + '_First_Name']
                data_dict[cur_npi]['Middle_Name'] = row['Principal_Investigator_' + str(i) + '_Middle_Name']
                data_dict[cur_npi]['Last_Name'] = row['Principal_Investigator_' + str(i) + '_Last_Name']
                data_dict[cur_npi]['Name_Suffix'] = row['Principal_Investigator_' + str(i) + '_Name_Suffix']
                data_dict[cur_npi]['Primary_Business_Street_Address_Line1'] = row['Principal_Investigator_' + str(i) + '_Business_Street_Address_Line1']
                data_dict[cur_npi]['Primary_Business_Street_Address_Line2'] = row['Principal_Investigator_' + str(i) + '_Business_Street_Address_Line2']
                data_dict[cur_npi]['City'] = row['Recipient_City']
                data_dict[cur_npi]['State'] = row['Recipient_State']
                data_dict[cur_npi]['Zip_Code'] = row['Recipient_Zip_Code']
                data_dict[cur_npi]['Country'] = row['Recipient_Country']
                data_dict[cur_npi]['Province'] = row['Recipient_Province']
                data_dict[cur_npi]['Postal_Code'] = row['Recipient_Postal_Code']
                for j in range(1,7):
                    data_dict[cur_npi]['Primary_Type_' + str(j)] = row['Principal_Investigator_' + str(i) + '_Primary_Type_'+ str(j)]
                for j in range(1,7):
                    data_dict[cur_npi]['Specialty_' + str(j)] = row['Principal_Investigator_' + str(i) + '_Specialty_' + str(j)]           
```

```python
pickle.dump(data_dict, open('./data/npi_info_dict.pkl', 'wb'))
```
